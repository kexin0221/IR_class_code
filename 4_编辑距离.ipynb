### 编辑距离

```python
word1 = "abcd"
word2 = "abcde"
```

```python
def editDistance(word1, word2):
    m, n = len(word1), len(word2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j], dp[i][j - 1], dp[i - 1][j - 1]) + 1

    return dp
```

```python
editDistance(word1, word2)
```

```python
print(f"编辑距离:{editDistance(word1, word2)[len(word1)][len(word2)]}")
```

```python
editDistance("input", "insert")
```

```python
print(f"编辑距离:{editDistance("input", "insert")[len("input")][len("insert")]}")
```
